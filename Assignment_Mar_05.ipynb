Q1. How can you create a Bokeh plot using Python code?

```python

# Importing necessary modules
from bokeh.plotting import figure, output_file, show

# Define the output file (HTML file where the plot will be saved)
output_file("line_plot.html")

# Create a new plot with a title and axis labels
p = figure(title="Simple Line Plot", x_axis_label='X-axis', y_axis_label='Y-axis')

# Define data points for the line plot
x_values = [1, 2, 3, 4, 5]
y_values = [2, 5, 8, 3, 7]

# Add a line renderer with data points
p.line(x_values, y_values, legend_label="Line", line_width=2)

# Show the plot (opens in a web browser)
show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

```python
'''In Bokeh, a glyph is a visual representation of your data. It is a fundamental concept that allows you to display data
points or shapes on a plot. Glyphs can represent various visual elements such as lines, circles, squares, bars, and more. 
By combining different glyphs, you can create complex and interactive visualizations.

To add glyphs to a Bokeh plot, you typically use specific glyph methods provided by the bokeh.plotting module, such as line(), 
circle(), square(), rect(), etc. Each of these methods allows you to specify properties like coordinates, sizes, colors, and
other visual attributes for the glyphs.

Here's an example that demonstrates how to create a Bokeh plot with multiple glyphs:

 Python script to generate a Bokeh plot with different glyphs:'''

# Import necessary modules
from bokeh.plotting import figure, output_file, show

# Define the output file (HTML file where the plot will be saved)
output_file("glyphs_example.html")

# Create a new plot with a title
p = figure(title="Glyphs Example", x_axis_label='X-axis', y_axis_label='Y-axis')

# Define data points for the glyphs
x_values = [1, 2, 3, 4, 5]
y_values = [2, 5, 8, 3, 7]

# Add glyphs to the plot
p.line(x_values, y_values, legend_label="Line", line_width=2, line_color="blue")
p.circle(x_values, y_values, legend_label="Circle", size=10, fill_color="red")
p.square(x_values, y_values, legend_label="Square", size=8, fill_color="green")

# Show the plot (opens in a web browser)
show(p)
'''When you run the above Python script, it will generate an HTML file named glyphs_example.html containing a Bokeh plot with
a line, circle, and square glyphs.
In this example:

The line() method adds a line glyph to the plot representing the data points (x_values, y_values).
The circle() method adds circle glyphs to the plot at the specified coordinates (x_values, y_values).
The square() method adds square glyphs to the plot at the specified coordinates (x_values, y_values).'''
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?


    Customizing the appearance of a Bokeh plot involves adjusting various components such as axes, title, legend, grid lines, 
    colors, fonts, and more. Bokeh provides a comprehensive set of options and properties to customize the visual attributes 
    of these components.

    Here's a guide demonstrating how to customize the appearance of a Bokeh plot, including the axes, title, and legend:

    # Import necessary modules
    from bokeh.plotting import figure, output_file, show

    # Define the output file (HTML file where the plot will be saved)
    output_file("customized_plot.html")

    # Create a new plot with custom properties
    p = figure(title="Customized Bokeh Plot",  # Set the title of the plot
               x_axis_label='X-axis Label',   # Set the label for the x-axis
               y_axis_label='Y-axis Label',   # Set the label for the y-axis
               plot_width=600,                 # Set the width of the plot
               plot_height=400,                # Set the height of the plot
               background_fill_color="lightgrey",  # Set the background color
               title_text_font_size="20pt",    # Set the font size for the title
               toolbar_location='above')        # Set the location of the toolbar

    # Define data points for the plot
    x_values = [1, 2, 3, 4, 5]
    y_values = [2, 5, 8, 3, 7]

    # Add glyphs to the plot
    p.line(x_values, y_values, legend_label="Line", line_width=2, line_color="blue")
    p.circle(x_values, y_values, legend_label="Circle", size=10, fill_color="red")

    # Customize axes properties
    p.xaxis.axis_label_text_font_size = "16pt"   # Set x-axis label font size
    p.yaxis.axis_label_text_font_size = "16pt"   # Set y-axis label font size
    p.xaxis.major_label_text_font_size = "12pt"  # Set x-axis tick label font size
    p.yaxis.major_label_text_font_size = "12pt"  # Set y-axis tick label font size

    # Customize legend properties
    p.legend.location = "top_left"               # Set legend location
    p.legend.label_text_font_size = "14pt"       # Set legend label font size

    # Show the plot (opens in a web browser)
    show(p)

    Explanation:
    Title and Labels: The title parameter is used to set the title of the plot, and x_axis_label and y_axis_label parameters are 
    used to set the labels for the x-axis and y-axis, respectively.
    Plot Dimensions and Background: The plot_width and plot_height parameters are used to set the dimensions (width and height) 
    of the plot. The background_fill_color parameter sets the background color of the plot.
    Font Sizes: You can customize the font size of the title, axes labels, and tick labels using properties like 
    title_text_font_size, axis_label_text_font_size, and major_label_text_font_size.
    Legend: The legend.location property is used to set the position of the legend on the plot, and legend.
    label_text_font_size is used to set the font size of the legend labels.


Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

    Using Flask:
    Flask is a lightweight Python web framework. Here's how you can embed a Bokeh plot into a Flask application:

    Install the necessary packages:
    pip install bokeh flask
    Create a Flask application (app.py) and embed the Bokeh plot:

    from flask import Flask, render_template
    from bokeh.plotting import figure, output_file, show
    from bokeh.embed import components

    app = Flask(__name__)

    @app.route('/')
    def index():
        # Create Bokeh plot
        p = figure(title="Embedded Bokeh Plot", x_axis_label='X-axis', y_axis_label='Y-axis')
        p.line([1, 2, 3, 4, 5], [2, 5, 8, 3, 7], line_width=2)

        # Get components of the plot (script and div)
        script, div = components(p)

        # Render the template with the script and div components
        return render_template('index.html', script=script, div=div)

    if __name__ == '__main__':
        app.run(debug=True)
    Create a Flask template (templates/index.html) to render the Bokeh plot:
    html
    Copy code
    <!DOCTYPE html>
    <html lang="en">
    <head>
        <meta charset="UTF-8">
        <title>Embedded Bokeh Plot</title>
        <!-- Include Bokeh CSS and JS files -->
        {{ script|safe }}
    </head>
    <body>
        <!-- Embed Bokeh plot -->
        {{ div|safe }}
    </body>
    </html>
    Run the Flask application:

    python app.py
    Visit http://127.0.0.1:5000/ in your web browser to see the embedded Bokeh plot.

    2. Using Django:
    Django is a high-level Python web framework that follows the Model-View-Controller (MVC) architecture. Here's how you can embed a Bokeh plot into a Django application:

    Create a new Django project and app:

    django-admin startproject myproject
    cd myproject
    python manage.py startapp myapp
    Install the necessary packages:

    pip install bokeh django
    Update the Django app (myapp/views.py) to embed the Bokeh plot:

    from django.shortcuts import render
    from bokeh.plotting import figure, output_file, show
    from bokeh.embed import components

    def bokeh_plot(request):
        # Create Bokeh plot
        p = figure(title="Embedded Bokeh Plot", x_axis_label='X-axis', y_axis_label='Y-axis')
        p.line([1, 2, 3, 4, 5], [2, 5, 8, 3, 7], line_width=2)

        # Get components of the plot (script and div)
        script, div = components(p)

        # Render the template with the script and div components
        return render(request, 'bokeh_plot.html', {'script': script, 'div': div})
    Create a Django template (myapp/templates/bokeh_plot.html) to render the Bokeh plot:
    html
    Copy code
    <!DOCTYPE html>
    <html lang="en">
    <head>
        <meta charset="UTF-8">
        <title>Embedded Bokeh Plot</title>
        <!-- Include Bokeh CSS and JS files -->
        {{ script|safe }}
    </head>
    <body>
        <!-- Embed Bokeh plot -->
        {{ div|safe }}
    </body>
    </html>
    Update the Django URLs (myproject/urls.py) to include the Bokeh plot view:
    from django.urls import path
    from myapp.views import bokeh_plot

    urlpatterns = [
        path('bokeh_plot/', bokeh_plot, name='bokeh_plot'),
    ]
    Run the Django development server:

    python manage.py runserver
    Visit http://127.0.0.1:8000/bokeh_plot/ in your web browser to see the embedded Bokeh plot.
